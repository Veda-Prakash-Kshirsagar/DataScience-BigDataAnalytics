```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
df.shape
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop("User ID",axis=1,inplace = True)
```

```python
df.head()
```

```python
df.drop("Gender",axis=1,inplace = True)
df.head()
```

```python
x = df.drop("Purchased",axis=1)
```

```python
x.head(2)
```

```python
y = df.Purchased
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
xtrain = sc.fit_transform(xtrain)
```

```python
xtest = sc.fit_transform(xtest)
```

```python
xtest
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
ytrain_predict = lr.predict(xtrain)
```

```python
ytest_predict = lr.predict(xtest)
```

```python
ytrain_predict
```

```python
ytest_predict
```

```python
print(lr.predict([[46,41000]]))
```

```python
lr.predict([[46,41000]])
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
matrix = confusion_matrix(ytest,ytest_predict)
print(matrix)
```

```python
score = accuracy_score(ytest,ytest_predict)
print(score)
```

```python
class_report = classification_report(ytest,ytest_predict)
print(class_report)
```

```python
import seaborn as sns
sns.heatmap(matrix,annot = True)
```

```python
tn , fp , fn , tp = confusion_matrix(ytest,ytest_predict).ravel()
```

```python
print(tn,fp,fn,tp)
```

```python
print("Error rate: ",(fp+fn)/(tn+fp+fn+tp))
```

```python
print("Accuracy: ",accuracy_score(ytest,ytest_predict))
```

```python
print("Precision: ",(tp)/(tp+fp))
```

```python
print("Recall: ",(tp+tn)/(tp+fp+tn+fn))
```
