```python
#Above error occurs because corr() works on numerical values only.
#Therefor applying label encoding to convert categorical data into numerical data
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dataset['sex'] = label_encoder.fit_transform(dataset['sex'])
dataset['sex'].unique()
```

```python
#for embarked
dataset['embarked'] = label_encoder.fit_transform(dataset['sex'])
dataset['embarked'].unique()
#for class
dataset['class'] = label_encoder.fit_transform(dataset['class'])
dataset['class'].unique()
#for who
dataset['who'] = label_encoder.fit_transform(dataset['who'])
dataset['who'].unique()
#foe adult_male
dataset['adult_male'] = label_encoder.fit_transform(dataset['adult_male'])
dataset['adult_male'].unique()
#for deck
dataset['deck'] = label_encoder.fit_transform(dataset['deck'])
dataset['deck'].unique()
#for embark_town
dataset['embark_town'] = label_encoder.fit_transform(dataset['embark_town'])
dataset['embark_town'].unique()
#for alive
dataset['alive'] = label_encoder.fit_transform(dataset['alive'])
dataset['alive'].unique()
#for alone
dataset['alone'] = label_encoder.fit_transform(dataset['alone'])
dataset['alone'].unique()
```
